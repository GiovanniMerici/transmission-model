```python
import pandas as pd
import numpy as np
from numpy.random import randint, rand
import matplotlib.pyplot as plt

class Person:
    def __init__(self):
        self.infected = False
        self.ss = randint(0,10)
        self.mask = randint(0,2)
        self.heal = 0
        self.fragility = randint(0,10)
        self.alive = 1
    def infection(self):
        self.infected = True
    def healed(self):
        self.infected = False
        self.mask = 1
    def death():
        self.alive = 0
        
   
        
N=500
p=0.5
k=5
```

```python
pop = [Person() for i in range(N)]
pop[0].infection()
pop[0].ss = 1
pop[0].mask = 0
```

```python
t=0
c=0
d=0
count= []
countd=[]
time=[]
while t < 100:    
    for i in range (N):
        if pop[i].alive == 1:
            if pop[i].infected == True:
                if pop[i].ss < 5:
                    if pop[i].mask == 0:
                        contacts = randint(N, size=k)
                        for j in contacts:
                            if rand() < p:
                                pop[j].infection()
                if pop[i].fragility < 8:
                    pop[i].heal = pop[i].heal + 1
                    if pop[i].heal > 7:
                    pop[i].healed()
            if pop[i].fragility > 8:
                pop[i].death()
                
    t = t +1
    print('time',t)
    c=0
    for i in range (N):
        if pop[i].infected == True:
            c=c+1
            
            
    count.append(c)        
    print(c)
    countd.append(N)
    time.append(t)
    plt.plot(time, count)
    plt.show()
    plt.plot(time, countd, color='red')
    plt.show()
    
    if 500 in count:
        break
print(count)
print('pop',N)
```
